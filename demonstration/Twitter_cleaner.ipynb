```python
from pandas import DataFrame
from pandas import read_excel
from pandas import read_csv
```

```python
df = read_excel('/home/martin/GitProjects/TAR-Project/evaluation/trump.xls')
```

```python
trump_tweets = []
for tweet in df['Unnamed: 3'][4:]:
    trump_tweets.append(tweet)
```

```python
trump_tweets
```

```python
from nltk.tokenize import TweetTokenizer
import re


def preprocess(tweet):
    tt = TweetTokenizer()
    text = tweet

    #uklanjanje hiperlinkova
    text = re.sub(r"http\S+", "URL", text)

    #uklanjanje brojeva
    text = re.sub(r"(\+|\-)*(\d+\.\d+)|\+(\d+)|\-(\d+)|(\d+)", "NUMBER", text)

    return tt.tokenize(text)
```

```python
preprocessed_tweets = [preprocess(tweet) for tweet in trump_tweets]
```

```python
import pickle

output_file = open("trump.pkl", "wb")
pickle.dump(preprocessed_tweets, output_file)
output_file.close()
```

```python
import pickle

def tweets_csv_to_pkl(fname):
    f = open(fname).readlines()
    firstLine = f.pop(0) #removes the first line
    lines = []
    for line in f:
        if line.strip():
            lines.append(line.strip())
    
    preprocessed = [preprocess(tweet) for tweet in lines]
    print(preprocessed[12])
    
    output_file = open(fname.replace("_tweets.csv", "") + ".pkl", "wb")
    pickle.dump(preprocessed, output_file)
    output_file.close()
```
